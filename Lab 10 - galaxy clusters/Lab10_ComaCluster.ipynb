## Step 2 - Velocity dispersion

Because the galaxies have motions in random directions (not necessarily in our line of sight), we can't directly measure the orbital speed of each galaxy. To find the total mass of the cluster, we can use a modified form of the enclosed mass equation:

$$M_{tot}=5\frac{ \langle v \rangle ^2 R}{G}$$

where the "5" accounts for the 3D shape of the cluster, R is the radius of the cluster, and G is the gravational constant.

The `<v>` term is the velocity dispersion, a measurement of the spread of different velocities of the galaxies in the cluster.
The galaxies themselves have random orbital speeds around the cluster, so some are slightly blueshifted and some are slightly redshifted from the mean. This causes the spread in the observed velocities, i.e. the "dispersion".
